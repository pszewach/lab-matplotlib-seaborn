#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.arange(0,100)
y1=x**2
y2=np.exp(x)

fig,(ax1,ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.set_title("Normal", fontweight=800)
ax2.plot(x,y1)
ax2.plot(x,y2)
ax2.set_yscale("log")
ax2.set_title("Logarithmic")
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python

fig,(ax1,ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.set_title("Normal", fontweight=800)
ax2.plot(x,y1)
ax2.plot(x,y2)
ax2.set_yscale("log")
ax2.set_title("Logarithmic")
ax1.legend([1, 2])
ax2.legend([1, 2])
plt.show()
```

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
x = ['Minutes Sedentary', 'Minutes Lightly Active','Minutes Very Active']
y= fitbit['Steps']
for variable in x:
    figure = plt.figure
    ax = plt.gca()
    ax.scatter(fitbit[variable], y)
    ax.set_xlabel(variable)
    ax.set_ylabel('Steps')
    ax.set_title(f"{variable} vs. Steps")
    plt.show()

#for i in range(len(plot_list)):
#    plt.figure()
#    plt.plot(plot_list[i])
#    plt.show()
```
